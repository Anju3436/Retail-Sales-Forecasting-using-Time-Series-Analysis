# **LOAD DATASET**

```python
df = pd.read_csv("sales data-set.csv")
df.head()
```

# **UNDERSTAND DATA**

```python
df.info()
df.describe()
```

# **CONVERT DATE COLUMN**

```python
df['Date'] = pd.to_datetime(df['Date'], dayfirst=True)
df = df.set_index('Date')
df.head()
```

```python
df = df.reset_index()
df.head()
```

# **HANDLE MISSING VALUES**

```python
df.isnull().sum()
df = df.fillna(method ='ffill')
```

# **Resample Monthly Sales**

```python
monthly = df['Weekly_Sales'].resample('M').sum()
monthly.head()
```

# **DATA VISUALIZATION**

# **Line Plot**

```python
plt.figure(figsize=(12,6))
plt.plot(monthly)
plt.title("Monthly Sales Trend")
plt.show()
```

# **Rolling Mean ( Trend)**

```python
rolling_mean = monthly.rolling(window=3).mean()
plt.plot(monthly, label='Original')
plt.plot(rolling_mean, label='Rolling Mean')
plt.legend()
plt.show()
```

# **Check Seasonality**

```python
from statsmodels.tsa.seasonal import seasonal_decompose

decompose = seasonal_decompose(monthly, model='additive')
decompose.plot()
plt.show()
```

# **Build ARIMA Model**

```python
from statsmodels.tsa.arima.model import ARIMA
model = ARIMA(monthly , order=(2,1,2))
model_fit = model.fit()
print(model_fit.summary())
```

# **FORECAST FUTURE SALES**

```python
forecast = model_fit.forecast(steps=12)
forecast
```

# **PLOT FORECAST**

```python
plt.figure(figsize=(10,5))
plt.plot(monthly, label='Actual Sales')
plt.plot(forecast, label='Forecast', color='red')
plt.legend()
plt.show()
```

# **PROPHET MODEL**

```python
from prophet import Prophet
df_prophet = monthly.reset_index()
df_prophet.columns = ['ds','y']
model = Prophet()
model.fit(df_prophet)
future = model.make_future_dataframe(periods=12,freq='M')
forecast = model.predict(future)
model.plot(forecast)
plt.show()
```
